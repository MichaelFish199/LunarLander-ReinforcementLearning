# Game Setup

```python
import warnings
warnings.simplefilter("ignore")
```

```python
# Import game environment
import gym
```

```python
# Environment
env = gym.make("LunarLander-v2")
```

```python
# Number of actions
env.action_space
```

```python
# Game with random actions
episodes = 5

for episode in range(episodes):
    state = env.reset()
    done = False
    
    while not done:
        new_state, reward, done, info = env.step(env.action_space.sample())
        env.render()
```

```python
# Closing environment
env.close()
```

# Preprocessing Environment

```python
state = env.reset()
print("dimensions =" ,state.shape)
```

```python
from stable_baselines3.common.vec_env import DummyVecEnv

# Wraping inside dummy environment
env = DummyVecEnv([lambda: env])

state = env.reset()
print("dimensions =" ,state.shape)
```

# Training The RL Model

```python
from stable_baselines3 import PPO

# Creating PPO model
model = PPO("MlpPolicy", env, verbose=1)
```

```python
model.learn(total_timesteps=100_000)
```

```python
model.save("best_model_iter200000")
```

# Testing The Model

```python
model = model.load("best_model_iter200000")
```

```python
# Starting the game
state = env.reset()
# Loop through the game
for episode in range(2000):
    action, _ = model.predict(state)
    state, reward, done, info = env.step(action)
    env.render()
    
```
